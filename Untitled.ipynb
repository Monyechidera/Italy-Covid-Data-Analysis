```python
#The code below converts the date column into a datetime column
covid_data['date'] = pd.to_datetime(covid_data.date)
```

```python
covid_data['year'] = pd.DatetimeIndex(covid_data.date).year
covid_data['month'] = pd.DatetimeIndex(covid_data.date).month
covid_data['day'] = pd.DatetimeIndex(covid_data.date).day
covid_data['weekday'] = pd.DatetimeIndex(covid_data.date).weekday
```

```python
covid_data
#so there is a new dataframe containing year, month, day and weekday columns
```

```python
# To query the table for July
covid_data_july = covid_data[covid_data.month == 7]

# To extreact some of the columns for further analysis
covid_data_july_metrics = covid_data_july[['new_cases', 'new_deaths', 'new_tests']]

# To get the sum for new cases, new deaths and new tests for July
covid_july_totals = covid_data_july_metrics.sum()
```

```python
covid_july_totals
```

```python
#we can group the data to compare the total figures by month
covid_month_data = covid_data.groupby('month')[['new_cases', 'new_deaths', 'new_tests']].sum()
```

```python
covid_month_data
```

```python
#we can do the same as above for the average figures
covid_month_mean_data = covid_data.groupby('month')[['new_cases', 'new_deaths', 'new_tests']].mean()
```

```python
covid_month_mean_data
```

```python
# To create a new data frame that would show the cumulative sum 
covid_data['total_cases'] = covid_data.new_cases.cumsum()
covid_data['total_deaths'] = covid_data.new_deaths.cumsum()
```
